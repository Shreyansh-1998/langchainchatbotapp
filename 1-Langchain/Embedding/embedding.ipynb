```python
embeddings_1024=OpenAIEmbeddings(model="text-embedding-3-large",dimensions=1024)
```

```python
embeddings_1024
```

```python
text="Hey, my name is Shreyansh Singh. Here I am creating a GenAI chatbot using OPENAI embeddings"
query_result=embeddings_1024.embed_query(text)
len(query_result)
```

```python
from langchain_community.document_loaders import TextLoader
loader=TextLoader("/Users/ssingh/Desktop/GenerativeAIApplications/Langchain/1-Langchain/Transformation/speech.txt")
docs=loader.load()
docs
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter=RecursiveCharacterTextSplitter(chunk_size=500,chunk_overlap=50)
final_documents=text_splitter.split_documents(docs)
final_documents
```

```python
#Embedding and Storing it in a vector database
from langchain_community.vectorstores import Chroma
db=Chroma.from_documents(final_documents,embeddings_1024)

db
```

```python
query="It will be all the easier for us to conduct ourselves as belligerents"
retrieve_results=db.similarity_search(query)
retrieve_results
```
